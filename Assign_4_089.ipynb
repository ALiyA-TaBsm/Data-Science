```python
#import warnings to 'ignore' warning messages

import warnings
warnings.filterwarnings("ignore")
```

```python
#import important libraries

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import manhattan_distances
from sklearn.metrics.pairwise import euclidean_distances

import pandas as pd
import math
```

```python
#generate two simple texts as examples

S1 = "data science is one of the most important courses in computer science"
S2 = "this is one of the best data science courses"
S3 = "the data scientists perform data analysis"
```

```python
#combine both texts to make a corpus

data = [S1, S2, S3]
data
```

```python
#create a reference object to the CountVectorizer constructor
#use the reference object to generate BoW matrix using the fit_transform() function

count_vectorizer = CountVectorizer()
c_vector_matrix = count_vectorizer.fit_transform(data)

#generate tf.idf matrix using TfidfVectorizer
tfidf_vect_exp = TfidfVectorizer()
tfidf_matrix_exp = tfidf_vect_exp.fit_transform(data)
```

```python
#use get_feature_names_out() to extract vocabulary

c_tokens_exp = count_vectorizer.get_feature_names_out()
c_tokens_exp
```

```python
#convert BoW matrix to an array format

vector_array = c_vector_matrix.toarray()
vector_array
```

```python
#convert BoW matrix to a Pandas dataframe

df_c_exp = pd.DataFrame(data = vector_array, columns = c_tokens_exp)
print("Dataframe of BoW matrix : ",)
df_c_exp
```

```python
#convert tf.idf matrix to a Pandas dataframe

df_t_exp = pd.DataFrame(data = tfidf_matrix_exp.toarray(), columns = c_tokens_exp)
print("Dataframe of tf.idf matrix : ")
df_t_exp
```

```python
#generate tf matrix
tf_matrix = vector_array / vector_array.sum(axis=1, keepdims=True)
print("Tf Matrix : \n",tf_matrix)

#convert tf matrix to a Pandas dataframe
df_tf_exp = pd.DataFrame(data = tf_matrix, columns = c_tokens_exp)
print("Dataframe of tf matrix : ")
df_tf_exp
```

```python
#sum of doc, words appear in.
num_data = len(vector_array)
num_terms = len(vector_array[0])
sum_list = []
for i in range(num_terms):
  count = 0
  for j in range(num_data):
    if(vector_array[j][i] > 0):
       count += 1
  sum_list.append(count)
sum_list
```

```python
#idf matrix
idf_matrix = []
for i in range(num_terms):
  idf_matrix.append(math.log10(num_data/sum_list[i]))
print("Idf matrix : \n",idf_matrix )


#convert idf matrix to a Pandas dataframe
df_idf_exp = pd.DataFrame(data = idf_matrix)
print("Dataframe of idf matrix : \n")
df_idf_exp

```

```python
#tf.idf matrix through manual calculate tf and idf matrix
tf_idf_matrix = {}
for i in range(num_terms):
  tf_idf_matrix = tf_matrix*idf_matrix
print("tf.idf matrix : \n",tf_idf_matrix )

#convert tf.idf matrix to a Pandas dataframe
df_tf_idf_exp = pd.DataFrame(data = tf_idf_matrix, columns = c_tokens_exp)
print("Dataframe of tf.idf matrix : ")
df_tf_idf_exp
```

```python
#generate similarity matrix
#Similarity matrixes of BoW vector matrix
c_cosine_similarity_matrix = cosine_similarity(c_vector_matrix)
vec_manhattan_dist = manhattan_distances(c_vector_matrix)
vec_euclidean_dist = euclidean_distances(c_vector_matrix)
print("\BoW vector Matrix: \n")

print("\nCosine Similarity:")
print(c_cosine_similarity_matrix)

print("\nManhattan Distance:")
print(vec_manhattan_dist)

print("\nEuclidean Distance:")
print(vec_euclidean_dist)
```

```python
#Similarity matrixes of tf.idf(through TfidfVectorizer) matrix
tfidf_cosine_similarity = cosine_similarity(tfidf_matrix_exp)
tfidf_manhattan_dist = manhattan_distances(tfidf_matrix_exp)
tfidf_euclidean_dist = euclidean_distances(tfidf_matrix_exp)
print("\ntf.idf Matrix: \n")

print("\nCosine Similarity:")
print(tfidf_cosine_similarity)

print("\nManhattan Distance:")
print(tfidf_manhattan_dist)

print("\nEuclidean Distance:")
print(tfidf_euclidean_dist)
```

```python
#Similarity matrixes of tf.idf matrix
tf_idf_cosine_similarity = cosine_similarity(tf_idf_matrix)
tf_idf_manhattan_dist = manhattan_distances(tf_idf_matrix)
tf_idf_euclidean_dist = euclidean_distances(tf_idf_matrix)
print("\ntf.idf Matrix: \n")

print("\nCosine Similarity:")
print(tf_idf_cosine_similarity)

print("\nManhattan Distance:")
print(tf_idf_manhattan_dist)

print("\nEuclidean Distance:")
print(tf_idf_euclidean_dist)
```
