```python
#import libraries

from sklearn import preprocessing
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score


#import ML evaluation metrics
from sklearn.model_selection import KFold, LeavePOut ,cross_val_score
from sklearn import metrics, model_selection
```

```python
#Accesssing Google Drive

from google.colab import drive
drive.mount('/content/drive')
```

```python
#read CSV (dataset) file

df_genderIdentify = pd.read_csv('/content/drive/MyDrive/Copy of gender-prediction.csv')
df_genderIdentify
```

```python
#Convert strings values into numbers
labels = preprocessing.LabelEncoder()

df_genderIdentify['gender'] = labels.fit_transform(df_genderIdentify['gender'])

df_genderIdentify['beard'] = labels.fit_transform(df_genderIdentify['beard'])

df_genderIdentify['hair_length'] = labels.fit_transform(df_genderIdentify['hair_length'])

df_genderIdentify['scarf'] = labels.fit_transform(df_genderIdentify['scarf'])

df_genderIdentify['eye_color'] = labels.fit_transform(df_genderIdentify['eye_color'])
```

```python
#seperate output and input

input = df_genderIdentify.drop('gender', axis=1)
output = df_genderIdentify['gender']
```

```python
folds = KFold(n_splits = 10)
p_out = LeavePOut(p = 1)
model_rfc = RandomForestClassifier(n_estimators = 5, max_depth=3, n_jobs=-1, random_state=42)
```

```python
scores_rfc = []
for train_index, test_index in folds.split(input):
    X_train, x_test ,Y_train, y_test = input.iloc[train_index], input.iloc[test_index] ,output.iloc[train_index], output.iloc[test_index]

    model = RandomForestClassifier()
    model.fit(X_train, Y_train)
    y_pred = model.predict(x_test)

    scores_rfc.append(f1_score(y_test, y_pred))
scores_rfc
```

```python
#f1 score
predict_folds = model_selection.cross_val_score(estimator=model_rfc,X=input,y=output,cv=folds)
predict_folds
```

```python
predict = model_selection.cross_val_score(estimator=model_rfc,X=input,y=output,cv=p_out)
print ("Cross Validation Scores-{}".format(predict))
print("Average Cross Validation score : {}".format (predict.mean()))
```
